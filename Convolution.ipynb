```python
import numpy as np
import scipy.signal as ssg
import matplotlib.pyplot as plt
```

```python
# noise sample
v_in = np.random.normal(size=1000)
```

```python
s1 = plt.magnitude_spectrum(v_in)
```

```python
# low pass filter
tau = 5
low_pass = np.ones(tau) / tau
```

```python
# filtered sample
v_out = np.convolve(v_in, low_pass)
```

```python
s2 = plt.magnitude_spectrum(v_out)
```

```python
# reconstructed sample
v_rec, remainder  = ssg.deconvolve(v_out, low_pass)
```

```python
np.mean(v_rec-v_in)
```
